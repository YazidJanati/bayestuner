```python
def getLabelsfromGT(X,ground_truth):
    truelabels = []
    Labels = {i: ground_truth[i] for i in range(len(ground_truth))}
    def getClosestCentroid(x,ground_truth):
        d = {np.linalg.norm(x-Labels[i]): i  \
            for i in range(len(ground_truth))}
        return d[min(d.keys())]
    for point in X:
        truelabels.append(getClosestCentroid(point,ground_truth))
    return truelabels
```

```python
import math
rastringin_ = lambda X : -(10*len(X) + sum([x**2 - 10*np.cos(2*math.pi*x) for x in X]))
dim = 2
bounds = [Bound([-5.12,5.12],"continuous") for i in range(dim)]
```

```python
from optimizer import LocalOptimizer

print(rastringin_([0,0]))
tuner = BayesTuner(objective = rastringin_, 
                   bounds = bounds,
                   num_iter = 70,
                   acquisition = lambda i : EI(i,None),
                   num_samples = 50,
                   alpha = 1e-1)

optimal = tuner.tune()                    
```

```python
fig300 = plt.figure()
plt.plot(tuner.past_hyper[:,0],tuner.past_hyper[:,1],'bo',markersize = 3)
plt.show()
```

```python
def ackley(X,a=20, b=0.2, c=2*np.pi):
    d           =len(X)
    sum_sq_term = -a * np.exp(-b * np.sqrt(sum(X[i]**2 for i in range(d))/d))
    cos_term    = -np.exp(sum(np.cos(c*X[i])/d for i in range(d)))
    return a + np.exp(1) + sum_sq_term + cos_term


ackley_ = lambda X : - ackley(X)
```

```python
dim = 2
bounds = [Bound([-10,10],"continuous") for i in range(dim)]

tuner1 = BayesTuner(objective = ackley_,
                    bounds = bounds,
                    num_iter = 40,
                    num_samples = 10,
                    alpha = 1e-1)

optimal_ = tuner1.tune()
print(optimal_)
print(ackley_(optimal_))
```

```python
fig100 = plt.figure()
plt.plot(tuner1.past_hyper[:,0],tuner1.past_hyper[:,1],'bo',markersize = 3)
```

```python


def adjusted_rand(X): 
    res = None
    try: 
        res = adjusted_rand_score(DBSCAN(eps=X[0], min_samples=X[1], 
                              metric="euclidean", metric_params=None, 
                              algorithm="auto", leaf_size=X[2], 
                              p=None, n_jobs=None).fit(df_).labels_,df.values.T[-1:][0])
    except:
        res = -1
    return res
objective = lambda X : adjusted_rand(X)

bounds = [Bound([0.40,0.60],"continuous"),Bound([2,200],"discrete"),Bound([2,200],"discrete")]

tuner = BayesTuner(objective = objective,
                   bounds = bounds,
                   num_iter = 30,
                   num_samples = 15)
```
